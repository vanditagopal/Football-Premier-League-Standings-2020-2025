```python
df = df[['intRank', 'intPoints', 'idTeam','strTeam',"intPlayed",'intWin','intLoss','intDraw','intGoalsAgainst','intGoalDifference','strForm']]
display(df.head())
```

```python
df = df.rename(columns={
    'intRank': 'Rank',
    'intPoints': 'Points',
    'idTeam': 'Team id',
    'strTeam':'Team Name',
    'intPlayed':'Played',
    'intWin':'Win',
    'intLoss':'Loss',
    'intDraw':'Draw',
    'intGoalsAgainst':'GoalsAgainst',
    'intGoalDifference':'Goal Difference',
    'strForm':'Form'
})

display(df.head())
```

```python
#df.info()
#print(df.dtypes)


# converting to neumeric data types

df['Rank'] = df['Rank'].astype(int)
df['Points'] = df['Points'].astype(int)
df['Team id'] = df['Team id'].astype(int)
df['Played'] = df['Played'].astype(int)
df['Win'] = df['Win'].astype(int)
df['Loss'] = df['Loss'].astype(int)
df['Draw'] = df['Draw'].astype(int)
df['GoalsAgainst'] = df['GoalsAgainst'].astype(int)
df['Goal Difference'] = df['Goal Difference'].astype(int)





```

Loading data 

```python
MYSQL_HOST=os.getenv("MYSQL_HOST")
MYSQL_PORT=os.getenv("MYSQL_PORT")
MYSQL_DATABASE=os.getenv("MYSQL_DATABASE")
MYSQL_USER=os.getenv("MYSQL_USER")
MYSQL_PASSWORD=os.getenv("MYSQL_PASSWORD")
```

```python
server_conn=connector.connect(
    host=MYSQL_HOST,
    port=MYSQL_PORT,
    user=MYSQL_USER,
    password=MYSQL_PASSWORD,
    connection_timeout=10,
    autocommit=False,
    raise_on_warnings=True
)

server_cur=server_conn.cursor()
print(f"[SUCESS]-Connected to MYSQl server")
```

```python
server_cur.close()
server_conn.close()
```

```python
db_connection=connector.connect(
    host=MYSQL_HOST,
    port=MYSQL_PORT,
    user=MYSQL_USER,
    password=MYSQL_PASSWORD,
    database=MYSQL_DATABASE,
    connection_timeout=10,
    autocommit=False,
    raise_on_warnings=True
)

cur=db_connection.cursor()
print(f"[SUCESS]-Connected to Database")
```

```python
sql_table="standings"
cur.execute("SHOW TABLES LIKE %s",(f"{sql_table}",))


if cur.fetchone() is None:
    raise SystemExit(f"Table {sql_table} is not found.")
else:
    print(f"This table {sql_table} exists")

```

```python
#Start the UPSERT OPERATION

table_cols=['Rank','Points','Team id','Team Name','Played','Win','Loss','Draw','GoalsAgainst','Goal Difference','Form']

standing_df=df[table_cols]
```

```python
standing_df

standings_records_tuples=standing_df.itertuples(index=False,name=None)

list_of_standings_records_tuples=list(standings_records_tuples)
```

```python
UPSERT_SQL=f"""
INSERT INTO {sql_table}
(`Rank`,Points,Team_id,Team_Name,Played,Win,Loss,Draw,Goals_Against,Goals_Difference,Form)
VALUES(%s,%s,%s,%s,%s,%s,%s,%s,%s,%s,%s) AS src
ON DUPLICATE KEY UPDATE
`Rank`=src.Rank,
Points=src.Points,
Team_id=src.Team_id,
Team_name=src.Team_Name,
Played=src.Played,
Win=src.Win,
Loss=src.Loss,
Draw=src.Draw,
Goals_Against=src.Goals_Against,
Goals_Difference=src.Goals_Difference,
Form=src.Form;

"""  
```

```python
no_of_rows_uploaded_to_mysql=len(list_of_standings_records_tuples)
```

```python
try:
    cur.executemany(UPSERT_SQL,list_of_standings_records_tuples)
    db_connection.commit()
    print(f"[SUCCESS]-Upsert attempted for {no_of_rows_uploaded_to_mysql} rows")
except Exception as e:
    print(f"[Failed]-Rollback initiated due to error {e}")

finally:
    cur.close()
    db_connection.close()
    print("Database Connection Closed")
```
